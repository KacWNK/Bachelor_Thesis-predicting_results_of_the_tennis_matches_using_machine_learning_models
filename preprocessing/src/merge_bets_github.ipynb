```python
import pandas as pd
```

```python
bets_2021_2023 = pd.read_csv("../data/bets_2021-2023.csv")
bets_2018_2020 = pd.read_csv("../data/bets_2018-2020.csv")
```

```python
github_2021_2023 = pd.read_csv("../data/github_2021-2023.csv")
github_2018_2020 = pd.read_csv("../data/github_2018-2020.csv")
```

```python
bets = pd.concat([bets_2018_2020, bets_2021_2023], ignore_index=True)        
```

```python
github = pd.concat([github_2018_2020, github_2021_2023], ignore_index=True)  
```

### Zobaczmy czy nie ma match_id które jest tylko w jednej ramce

```python
bets[~bets['match_id'].isin(github['match_id'])]['match_id'].unique()
```

```python
github[~github['match_id'].isin(bets['match_id'])]['match_id'].unique()
```

### Możemy zatem zmergować bets z github za pomocą match_id

```python
matches = pd.merge(bets, github, on='match_id', how='inner')
```

```python
matches
```

```python
matches = matches.drop(columns=['ATP', 'Winner', 'Loser', 'surface', 'tourney_location_y', 'tourney_level', 'match_num', 'winner_id_y', 'loser_id_y', 'Best of', 'round', 'shortened_winner_name', 'shortened_loser_name','B365W','B365L','PSW', 'PSL','MaxW', 'MaxL', 'score', 'WRank',
       'LRank', 'WPts', 'LPts'], axis=1)
```

```python
matches = matches.rename(columns={'tourney_location_x': 'tournament_location', 'loser_id_x': 'loser_id', 'winner_id_x': 'winner_id', 'Tournament': 'tournament_name', 'Court': 'indoor_or_outdoor', 'Series': 'tournament_level', 'tourney_date': 'tournament_date', 'tourney_id': 'tournament_id'})
```

```python
matches.columns
```

```python
matches['Date'] = pd.to_datetime(matches['Date'], errors='coerce')
matches['tournament_date'] = pd.to_datetime(matches['tournament_date'], format='%Y%m%d', errors='coerce')
matches
```

```python
matches.to_csv("../data/merged_matches.csv", index=False)
```
